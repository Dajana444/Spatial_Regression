```
import mapclassify
import geopandas as gpd
import matplotlib.pyplot as plt
import holoviews.plotting as hvplot
```

```
relative_risk = pd.merge(geodata, relative_risk, on="kommunenummer")
relative_risk
```

```
f, ax = plt.subplots(1, figsize=(12, 12))
relative_risk.plot(column='SMR', scheme="Quantiles", k=5, ax=ax,
        edgecolor='grey', legend=True, cmap = "Reds", linewidth=0.3)
ax.set_axis_off()
ax.set_title('Standard incidence rate map for digestive cancer')
plt.show()
```

```
f, ax = plt.subplots(1, figsize=(12, 12))
relative_risk.plot(column='RR', scheme="Quantiles", k=5, ax=ax,
        edgecolor='grey', legend=True, cmap = "Reds", linewidth=0.3)
ax.set_axis_off()
ax.set_title('Reative risk map for digestive cancer')
plt.show()
```

```
relative_risk2 = pd.read_csv("https://raw.githubusercontent.com/Dajana444/Spatial_Regression/main/map5.csv", encoding= 'unicode_escape')
relative_risk2 = relative_risk2[["kommunenummer","SMR","RR"]]
relative_risk2
```

```
relative_risk2 = pd.merge(geodata, relative_risk2, on="kommunenummer")
relative_risk2
```

```
f, ax = plt.subplots(1, figsize=(12, 12))
relative_risk2.plot(column='SMR', scheme="FisherJenks", k=4, ax=ax,
        edgecolor='grey', legend=True, cmap = "Reds", linewidth=0.3)
ax.set_axis_off()
ax.set_title('The standard incidence rate mapping for lung cancer')
plt.show()
```

```
f, ax = plt.subplots(1, figsize=(12, 12))
relative_risk2.plot(column='RR', scheme="FisherJenks", k=4, ax=ax,
        edgecolor='grey', legend=True, cmap = "Reds", linewidth=0.3)
ax.set_axis_off()
ax.set_title('The relative risk map for lung cancer')
plt.show()
```
